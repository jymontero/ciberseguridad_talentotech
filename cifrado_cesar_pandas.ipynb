```python
import pandas as pd
import string
import requests
from datetime import datetime
import matplotlib.pyplot as plt

alfabeto = "abcdefghijklmnopqrstuvwxyz"
lista_frases = ["hola como estas","mensaje de cesar", "bootcam ciberseguridad", "mensaje descifrado", "cifrado cesar pandas"]
lista_saltos = [1, 5, 7, 8, 10]

data_dict = {
    "msj_original": lista_frases,
    "clave": lista_saltos
}

df = pd.DataFrame(data_dict)
#print(df)

def cifrar_cesar(texto, salto):
  resultado = ""
  for caracter in texto:
    if caracter.lower() in alfabeto:
      es_mayus = caracter.isupper()
      indice = alfabeto.index(caracter.lower())
      nuevo_indice = (indice + salto) % len(alfabeto)
      nuevo_caracter = alfabeto[nuevo_indice]
      resultado += nuevo_caracter.upper() if es_mayus else nuevo_caracter
    else:
      resultado += caracter
  return resultado

def descifrar_cesar(texto, salto):
    return cifrar_cesar(texto, -salto)

df["msj_cifrado"] = df.apply(lambda row: cifrar_cesar(row["msj_original"], row["clave"]), axis=1)
df["msj_descifrado"] = df.apply(lambda row: descifrar_cesar(row["msj_cifrado"], row["clave"]), axis=1)


df
```
